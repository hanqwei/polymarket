```python
from py_clob_client.client import ClobClient
from py_clob_client.clob_types import ApiCreds
import os
```

```python
host = "https://clob.polymarket.com"
chain_id = 137 # Polygon mainnet
private_key = os.getenv("PRIVATE_KEY")
api_creds = ApiCreds(os.getenv("API_KEY"), os.getenv("API_SECRET"), os.getenv("API_PASSPHRASE"))

client = ClobClient(
    host=host,
    chain_id=chain_id,
    key=private_key,  # Signer enables L1 methods
    signature_type=2,
    creds=api_creds  # API credentials enable authenticated endpoints
)
```
